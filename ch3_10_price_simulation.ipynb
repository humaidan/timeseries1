```python
import numpy as np
import matplotlib.pyplot as plt

T = 1000  # number of time steps
P0 = 10   # initial price
mu = 0.001  # drift
```

```python
last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
        # sample a log return
        r = 0.01 * np.random.randn()

        #computer the new log price
        p = last_p + mu + r

        # store the return and price
        log_returns[t] = r + mu
        prices[t] = np.exp(p)

        # assign last_p
        last_p = p

```

```python
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title('Simulated Prices')
```
